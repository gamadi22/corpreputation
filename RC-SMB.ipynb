# Reputación Corporativa

## Criterios y subcriterios

**1. Capability  (CP):**
- PP = Short and medium term profit projection (%). Mediremos el crecimiento con una variable numérica. El crecimiento puede ser positivo o negativo.
- IT = Innovation. La capacidad de innovación se medirá en un rango lingüístico en s5.
- BQ = Quality of business management. La capacidad y calidad en la gestión empresarial se medirá en un rango lingüístico en s5.

**2. Benevolence (BV):**
- RC = Respect for consumer rights. El respeto a los derechos de los consumidores se medirá en un rango lingüístico en s5. 
- PS = Quality of product-service. La calidad del producto / servicio se medirá en un rango lingüístico en s5.
- LQ = Labor quality. La calidad de la mano de obra se medirá en un rango lingüístico en s5.

**3. Integrity (INT):**
- EC = Ethical Commitment. El compromiso ético se medirá en un rango lingüístico en s3.
- SC = Social Commitment. El compromiso social se medirá en un rango lingüístico en s3.
- NC = Environmental Commitment. El compromiso con el entorno se medirá en un rango lingüístico en s5.

**4. Net Promote Score (NPS):**
- CR = Customer rating. La valoración del cliente se medirá en un rango lingüístico en s5.
- SA = Supplier assessment. La valoración de colaboradores externos se medirá en un rango lingüístico en s5.
- IA = Internal assessment. La valoración interna se medirá en un rango lingüístico en s5.

### Empresas objetivo de estudio

Se escogen 5 empresas proveedores de servicios de tecnología, nuestro objetivo principal es obtener una valoración de cada empresa de acuerdo con la puntuación obtenida en el proceso de scoring reputación corporativa.

Definimos un dataframe con los subcriterios y las valoraciones de cada empresa para cada subcriterio.
En un segundo dataframe incorporaremos los criterios, se obtendrán como evaluación global de cada criterio.

```python
# Definir los posibles valores para cada variable
# PP, IT, BQ, RC, PS, LQ, EC, SC, NC, CR, SA, IA
import random
import pandas as pd
import numpy as np

pp_values = list(range(0, 41, 5))
it_values = [0, 1, 2, 3, 4]
bq_values = [0, 1, 2, 3, 4]
rc_values = [0, 1, 2, 3, 4]
ps_values = [0, 1, 2, 3, 4]
lq_values = [0, 1, 2, 3, 4]
ec_values = [0, 1, 2, 3, 4]
sc_values = [0, 1, 2, 3, 4]
nc_values = [0, 1, 2, 3, 4]
cr_values = [0, 1, 2, 3, 4]
sa_values = [0, 1, 2, 3, 4]
ia_values = [0, 1, 2, 3, 4]

# Definir las probabilidades para cada valor
pp_probabilities = [1] * 9  # Probabilidades uniformes ya que hay 9 posibles valores
it_probabilities = [0.10, 0.20, 0.30, 0.30, 0.10]
bq_probabilities = [0.10, 0.20, 0.30, 0.30, 0.10]
rc_probabilities = [0.10, 0.20, 0.40, 0.30, 0.10]
ps_probabilities = [0.10, 0.20, 0.40, 0.30, 0.10]
lq_probabilities = [0.10, 0.20, 0.40, 0.30, 0.10]
ec_probabilities = [0.10, 0.20, 0.40, 0.30, 0.10]
sc_probabilities = [0.10, 0.20, 0.40, 0.30, 0.10]
nc_probabilities = [0.10, 0.20, 0.40, 0.30, 0.10]
cr_probabilities = [0.10, 0.20, 0.40, 0.30, 0.10]
sa_probabilities = [0.10, 0.20, 0.40, 0.30, 0.10]
ia_probabilities = [0.10, 0.20, 0.40, 0.30, 0.10]


n = 100

# Crear una instancia de Random con la semilla
random_generator = random.Random(123)

pp_sample = random_generator.choices(pp_values, pp_probabilities, k=n)
it_sample = random_generator.choices(it_values, it_probabilities, k=n)
bq_sample = random_generator.choices(bq_values, bq_probabilities, k=n)
rc_sample = random_generator.choices(rc_values, rc_probabilities, k=n)
ps_sample = random_generator.choices(ps_values, ps_probabilities, k=n)
lq_sample = random_generator.choices(lq_values, lq_probabilities, k=n)
ec_sample = random_generator.choices(ec_values, ec_probabilities, k=n)
sc_sample = random_generator.choices(sc_values, sc_probabilities, k=n)
nc_sample = random_generator.choices(nc_values, nc_probabilities, k=n)
cr_sample = random_generator.choices(cr_values, cr_probabilities, k=n)
sa_sample = random_generator.choices(sa_values, sa_probabilities, k=n)
ia_sample = random_generator.choices(ia_values, ia_probabilities, k=n)


subcategories=['PP', 'IT', 'BQ', 'RC', 'PS', 'LQ', 'EC', 'SC', 'NC', 'CR', 'SA', 'IA']

# Crear el DataFrame
data = {'PP': pp_sample, 'IT': it_sample, 'BQ': bq_sample, 'RC': rc_sample, 'PS': ps_sample, 
       'LQ': lq_sample, 'EC': ec_sample, 'SC': sc_sample, 'NC': nc_sample, 'CR': cr_sample,
       'SA': sa_sample, 'IA': ia_sample}

df = pd.DataFrame(data)

# Imprimir el DataFrame
df
```

```python
def direct_2ts5(value_input):
    ranking_label = int(round(value_input*4))
    margen = value_input-ranking_label/4
    S = ("VL","L","M","H","VH")
    s_label = S[ranking_label]
       
    if margen==0:
        value_label=s_label
    else:
        value_label="("+s_label+","+str(round(margen,3))+")"
    
    return (value_label)
```

```python
# Buscará el número óptimo de clústeres basándose en el método del codo como se indica a continuación:
SS_distance = {}
for k in range(1,20):
    mod = KMeans(n_clusters= k, max_iter=1000,init = 'k-means++', random_state=123)
    mod = mod.fit(df4)
    SS_distance[k] = mod.inertia_

#Plotting the sum of square distance values and numbers of clusters
plt.figure(figsize = (15,5))
sns.pointplot(x = list(SS_distance.keys()), y = list(SS_distance.values()))
plt.xlabel("Número clusters")
plt.ylabel("Suma de distancias al cuadrado")
plt.title("Técnica del codo para encontrar el tamaño óptimo de las agrupaciones")
```

```python
from sklearn.metrics import silhouette_score
for n_cluster in range(2, 9):
    kmeans = KMeans(n_clusters=n_cluster, max_iter=1000, init = 'k-means++', random_state=123).fit(df4)
    label = kmeans.labels_
    sil_coeff = silhouette_score(df4, label, metric='euclidean')
    print("For n_clusters={}, The Silhouette Coefficient is {}".format(n_cluster, sil_coeff))

# Correlación entre variables
```
